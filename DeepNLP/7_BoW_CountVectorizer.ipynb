```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
allsentence=["This pasta is very tasty and affordable",
             "This pasta is not very tasty and is affordable",
             "This pasta is very very delicious"]
```

```python
CountVectorizer=CountVectorizer()
```

```python
x=CountVectorizer.fit_transform(allsentence)
```

```python
result = x.toarray()

print(result)
```

```python
text = ['Hello my name is james',
'james this is my python notebook',
'james trying to create a big dataset',
'james of words to try differnt',
'features of count vectorizer']
```

```python
import pandas as pd
CountVectorizer=CountVectorizer()
x=CountVectorizer.fit_transform(text)
result = x.toarray()
df=pd.DataFrame(data=result,columns=CountVectorizer.get_feature_names())
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
text=["hello my name is dhanya","Hello my name is dhanya"]
CountVectorizer=CountVectorizer(lowercase=False)
x=CountVectorizer.fit_transform(text)
result = x.toarray()
df=pd.DataFrame(data=result,columns=CountVectorizer.get_feature_names())
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
text=["hello my name is dhanya","Hello my name is dhanya"]
CountVectorizer=CountVectorizer(stop_words=['is','my']#'english')
x=CountVectorizer.fit_transform(text)
result = x.toarray()
df=pd.DataFrame(data=result,columns=CountVectorizer.get_feature_names())
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
text = ['This is the first document.','This document is the second document.','And this is the third one.', 'Is this the first document?',]
CountVectorizer=CountVectorizer(max_features=3)
x=CountVectorizer.fit_transform(text)
result = x.toarray()
df=pd.DataFrame(data=result,columns=CountVectorizer.get_feature_names())
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
text = ['This is the first document. Is this the first document?' ]
CountVectorizer=CountVectorizer(binary=True)
x=CountVectorizer.fit_transform(text)
result = x.toarray()
df=pd.DataFrame(data=result,columns=CountVectorizer.get_feature_names())
df
```
